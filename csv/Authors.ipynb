```python
import pandas as pd
import numpy as np
```

```python
article_df = pd.read_csv("../database/labelling.csv")
```

```python
article_df
```

```python
authors = article_df["Corresponding Author"]
```

```python
for key, value in authors.iteritems(): 
    print(key, value) 
    print() 
```

```python
for key, value in authors.iteritems(): 
    print(type(value))
    print() 
```

```python
a = set()
```

```python
for key, value in authors.iteritems(): 
    x = value.split(",")
    for element in x:
        a.add(element.strip())
```

```python
a
```

```python
authors = pd.DataFrame(list(a))
```

```python
authors.columns = ['name']
```

```python
authors
```

```python
authors.to_csv(index=False)
```
